Below we present a simple example of using **Net Present Value (NPV) calculations** in assessing ecological engineering technologies.  Our example assess NPV for a rainwater collection system in the Pacific Northwest.


## What is Net Present Value (NPV)?

Engineering expressly balances benefits and costs.  One metric of this balance is what is known as the “Net Present Value” of an undertaking.  The net present value expresses **the lifetime cost and benefit in terms of today’s value**.  Costs and benefits that are expected in out years are discounted in value by a **discount rate** that reflects the fact that potential future income has less value to you today than dollars you are given immediately.  Thus, if I offer you \\$1 today, and you indicate that you would equally accept \\$1.05 a year from today, then you are saying that your discount rate is 5%.

Given the **number of periods of interest** ($\large N $) (generally the useful lifetime of a project in years), an **annual discount rate** ($\large i$), and **annual costs** ($\large R_t$) (initial investment or maintenance), Net Present Value can be calculated as:

$\displaystyle \large {npv}(i,N)= \sum_{t=1}^N \frac{R_t}{(1+i)^t} $  


## Example: Calculating Net Present Values of a Roof Water Catchment System for Rainwater Capture

To make these ideas concrete, consider a roof water catchment system that people would use for displacing use of city water to irrigate their household plantings.  Our goal is to assess the NPV of a system, considering the initial install costs, maintenance costs, and returnsm given a discount rate and time frame for which costs will be assessed.

#### We will assume the following:

1. The design capacity of the Roof Water Catchment System is 5m<sup>3</sup>
2. The system has an **installation cost** of \\$500 per cubic meter of water storage capacity,
2. The system has **annual maintenance costs** that are 4\% of the initial cost,
3. Our **discount rate** is 5\%/year,
4. The **cost of the water** we are avoiding use of (the revenue value) is \\$5/m<sup>3</sup>/year.
5. The **life of the system** is 15 years
6. **Disposal cost** at the end of the system’s life is 10% of the installation cost.

#### A Python program for calculating NPV is given below.

The program below utilizes the publically available `ecoe` package.  Alternatively, you can [download an Excel spreadsheet]("NPV-example-rainwater-collection.xlsx") with the same NPV calculations.

```python
import ecoe.econ as ee
import matplotlib.pyplot as plt
from IPython.display import display

# basic problem parameters
capacity = 5             # m3
installRate = 500        # $/m3
maintRate = 0.05         # percent of installation costs/year
disposalRate = 50        # $/m3
discountRate = 0.05      # decimal percent
revenueRate = 5          # $/m3
annualWaterSavings = 75  # m3
lifetime = 15            # years|

# compute costs/returns
installCost   = installRate*capacity  # $ (first year only)
maintCost     = maintRate*installCost # $/year (all years)
disposalCost  = disposalRate*capacity # $ (last year only)
annualRevenue = revenueRate * annualWaterSavings   # $/year

# Generate an NPV table
df = ee.GetNPVTable(installCost,maintCost,disposalCost,annualRevenue,discountRate,lifetime)
display(df)

# get the payback period as well
pp = ee.GetDiscountedPaybackPeriod(discountRate, df['Net Cash Flow'].values)

# plot NPV, etc over time
plt.figure(figsize=(12,5))
plt.plot(df['Year'],df["Period Cost"],label="Annual Cost")
plt.plot(df['Year'],df["Period Revenue"],label="Annual Revenue")
plt.plot(df['Year'],df["Period NPV"],label="Period NPV")
plt.plot(df['Year'],df["Cumulative NPV"],label="Cumulative NPV", linewidth=6)
plt.plot( [pp,pp],[-1000,1200],linestyle='--')
plt.text(pp+0.2,-900, "Break even point")
plt.legend()
plt.xlabel("Years")
plt.ylabel("$")
plt.grid()
plt.show()
```

## Interpreting NPV in the Example Problem Above

What NPV tells us is the value of your investment at some point in the future, considering the "time value" of your money, compared to the alternative of investing your money now at the discount rate.

In the example above, a key point is that if this system were put in, say, Portland Oregon, because of the Mediterranean climate, the system would only fill about 5 times during the period in which irrigation was needed (reflected in the example above).  For a 5 cubic meter system, this leads to a NPV of -\\$2,495, so the system cost us much more than it saved!  In contrast, if you were in a mid-continental location, you might have it fill 15 times per irrigation season, yielding a positive \\$350 NPV.  In the unlikely event that you could use the capacity of the tank 25 times per year, the NPV would be nearly \\$3,200.  This sort of exploration of cost and benefit, which you could carry out for any of the assumed design values found above, illustrates how NPV allows you to readily get a feel for the likely net benefit of an intervention.  

It is worth thinking about some projects which only have expenses, without any income.  A pollution reducing system might have this character for the industry using that device.  Then the NPV will be only negative.  That said, it can still be very useful both to get a feel for the costs involved, and to compare different options, which might have differing initial, maintenance, and disposal costs. 

# Discounted Payback Period

The **discounted payback period** is method for determining the payback period for an investment that accounts for the time value of money. It indicates the amount of time that it will take for a project to “break even”, or to get the point where the net cash flows generated cover the initial cost of the project.  It can be used to evaluate the profitability and feasibility of a specific project over a given lifetime.

To calculate the discounted payback period, we discount the net cash flows and subtract them from the initial cost.  The point in time in which this is zero is the discounted payback period.   It is calculated using the `ecoe.GetDiscountedPaybackPeriod()` function in the code block above, and shown on the chart above.
