Problem 1 - using the data at http://explorer.bee.oregonstate.edu/Topic/Modeling/Data/WatershedFlow.csv, write a Python program to: 

1. retrieve the data
2. Print a list of column labels
3. Plot Summer Stream Flow (cfs) as a function of Drainage area (square miles)


```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel("http://explorer.bee.oregonstate.edu/Topic/Modeling/Data/OrCoastalStreams.xlsx")
print( df.columns)

plt.plot(df['DrainageArea_sqmi'], df['MeanAnnualDischarge_cfs'], '+')
plt.show()
```

Problem 2 - develop a linear regression for estimating Summer Stream Flow (cfs) based on the Drainage area (square miles).  Report the best-fit parameters and r2 (goodness of fit) value for the resulting model

```python
from scipy.stats import linregress

x = df['DrainageArea_sqmi']
y = df['MeanAnnualDischarge_cfs']


slope, intercept, r_value, p_value, std_err = linregress(x,y)
print( "linregress() results: Slope: {:.3}, Intercept: {:.3}, r2-value: {:.3}".format(slope, intercept, r_value**2 ) )

# generate modeled results
yModeled = slope * x + intercept

plt.plot(x,y,'o', color='blue' )   # add observation series
plt.plot(x, yModeled, '-', color='red' )    # add modeled series
plt.title( "Regression Results - Site 1")
plt.legend( ['Observed', 'Predicted'], loc='upper right')
plt.xlabel( "Drainage Area (sq mi)")
plt.ylabel( "Summer stream flow (CFS)")
plt.text( 130, 0.5, "r2={:.3}".format( r_value**2 ), fontsize=14 )

plt.show()
```
