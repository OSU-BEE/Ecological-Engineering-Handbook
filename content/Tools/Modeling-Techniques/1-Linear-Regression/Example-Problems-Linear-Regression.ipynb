## Problem 1

The first problem we will look at involves the ETRef dataset capturing Evapotranspiration fractions for two sites, collected daily over the course of a year.  A copy of the dataset is available at http://explorer.bee.oregonstate.edu/Topic/Modeling/Data/ETrF.xlsx.  Our goal will be to develop a linear regression for this data for one of the sites, reporting regression results and plotting the results 

Breaking this down into individual steps:

1) Get the data
2) Run the regression
3) Calculate and report statistics
4) Generate plots

For this exercise, we will use the `polyfit()` function in the **numpy** package to conduct the regression.

```python
# Step 1 - get the data
import numpy as np     # because it's numpy
import pandas as pd    # data library
import matplotlib.pyplot as plt

df = pd.read_excel( "http://explorer.bee.oregonstate.edu/Topic/Modeling/Data/ETrF.xlsx" )
#print( df )   # print the column headers to the console

plt.plot( df['Day'].values,df['Site1'].values, "+", color="green")
plt.plot( df['Day'].values,df['Site2'].values, "+")
plt.show()
```

Next, run the regression with `polyfit()`.  Because `polyfit()` does not calculate regression stats, we will have to do that on our own by generated modeled results for our data points and calcuating an $r^2$ from the differences between the observations and the modeled results.

```python
# Step 2 - run regression
xDays = df['Day']
ySite2 = df['Site2']

coefficients  = np.polyfit(xDays, ySite2, 2)    # solve the first order polynomial - returns list of coefficients

# calculate additional stats  For this, we will need to get some modeled results.
f = np.poly1d( coefficients )    # make a function that contains the regressed model
yModeled = f( xDays )            # apply the model to the observed days

# compute the correlation coefficient
yMean = np.mean( ySite2 )
sstot = np.sum( (ySite2-yMean)**2 )
ssr = np.sum( (ySite2-yModeled)**2 )
r2 = 1 - ssr/sstot

print( "Regressed Function: y=", f )
print( "polyfit: Slope: {:.3},   Intercept: {:.3},   R-squared: {:.3}".format( coefficients[0], coefficients[1], r2 ) )
```

Finally, let's generate a plot of the results.  For this example, we will plot the observed data as points, the modeled data as a line, and plot both vs the day of the year.

```python
import matplotlib.pyplot as plt

plt.plot(xDays,ySite2,'o', color='blue' )   # add observation series
plt.plot(xDays, yModeled, '-', color='red' )    # add modeled series
plt.title( "Regression Results - Site 2")
plt.legend( ['Observed', 'Predicted'], loc='upper right')
plt.xlabel( "Day of Year")
plt.ylabel( "ET Fraction")
plt.text( 130, 1.4, "r2={:.3}".format( r2 ), fontsize=14 )

plt.show()
```

## Problem 2

Next, we will use the same dataset to generate a regression using the `linregress()` function in the **scipy.stats** package.   This function is limited to first order polynomials (lines), but has the advantage of returning the regression stats back as part of the function. This time we will use the first site's data.  The steps are similar to what we did above, so we will combine things into a single code block.  Because we are using the same dataset as before we can utilize the DataFrame (df) we got above containing our dataset.

```python
from scipy import stats

ySite1 = df['Site1']    # get column from dataframe

# run regression
slope, intercept, r_value, p_value, std_err = stats.linregress(xDays,ySite1)
print( "linregress() results: Slope: {:.3}, Intercept: {:.3}, r2-value: {:.3}".format(slope, intercept, r_value**2 ) )

# generate modeled results
yModeled = slope * xDays + intercept

plt.plot(xDays,ySite1,'o', color='blue' )   # add observation series
plt.plot(xDays, yModeled, '-', color='red' )    # add modeled series
plt.title( "Regression Results - Site 1")
plt.legend( ['Observed', 'Predicted'], loc='upper right')
plt.xlabel( "Day of Year")
plt.ylabel( "ET Fraction")
plt.text( 130, 0.5, "r2={:.3}".format( r_value**2 ), fontsize=14 )

plt.show()
```

## Problem 3 - Predicting planting dates

The next problem involves predicting trends in planting dates for strawberries over time at a location near Corvallis, Oregon.  The dataset we will use is from [Agrimet](https://www.usbr.gov/pn/agrimet/), an agricultural weather reporting system with stations throughout the Pacific Northwest, and is available at http://explorer.bee.oregonstate.edu/Topic/Modeling/Data/crvo_1991_2016_plantdates.xlsx. It contains the observed planting dates for a number of crops in Corvallis for the period 1991-2016.  Teh planting dates for **strawberries** is given in the column "SBRY". The procedure will be similar to what we did above.  We will use the scipy.stats pack for the regression.

For plotting, we will generate a figure with two plots. The first will be similar to what we did above.  The second will be an "Line of Perfect Fit" plot, plotting predicted vs observed values, and a 1:1 perfect fit line. 

```python
import numpy as np        # because it's numpy
import pandas as pd       # data library
from scipy import stats   # for linregress()

# Step 1 - get the data
df = pd.read_excel( "http://explorer.bee.oregonstate.edu/Topic/Modeling/Data/crvo_1991_2016_plantdates.xlsx" )
print( df.columns.values )   # print the column headers to the console

xObs = df['Year']
yObs = df['SBRY']

# run regression
slope, intercept, r_value, p_value, std_err = stats.linregress(xObs,yObs)
print( "linregress() results: Slope: {:.3}, Intercept: {:.3}, r2-value: {:.3}".format(slope, intercept, r_value**2 ) )

# generate modeled results
yModeled = slope * xObs + intercept

# make a figure with two plots - first is observed, predicted vs. time
plt.figure(figsize=(16, 6))
plt.subplot( 1,2, 1)   
plt.plot(xObs,yObs,'o', color='blue' )   # add observation series
plt.plot(xObs, yModeled, '-', color='red' )    # add modeled series
plt.title( "Regression Results - Strawberry Planting Dates ")
plt.legend( ['Observed', 'Predicted'], loc='upper right')
plt.xlabel( "Year")
plt.ylabel( "Planting Day of Year")
plt.text( 1995, 58, "r2={:.3}".format( r_value**2 ), fontsize=14 )

# second is predicted vs observed
plt.subplot( 1, 2, 2)   
plt.plot(yModeled,yObs,'o', color='blue' )   # add observation series
plt.plot([55,75],[55,75], '-', color='red' )    # add modeled series
plt.title( "Regression Results - Strawberry Planting Dates ")
plt.legend( ['Data', 'Line of Perfect Fit'], loc='lower right')
plt.xlabel( "Observed Planting Date")
plt.ylabel( "Predicted Planting Date")

plt.show()

# now let's use the model to predict the future
yMean = np.mean( yObs )
y2040 = slope * 2040 + intercept
print( "Mean for period 1991-2016: {},  Prediction for 2040 is {}".format( int(yMean), int(y2040)))
```

## Estimating Fish Growth from Age, Temperature

For this problem we will look at involves estimating effect of age and temperature on fish lengths using the dataset is available at https://people.sc.fsu.edu/~jburkardt/datasets/regression/x06.txt.  Our goal will be to develop a linear regression for this data, reporting regression results and plotting the results 

As usual, we will break this down into teh following steps:

1. Get the data
2. Generate plots
3. Run the regression
4. Calculate and report statisticsk
5. Generate plots comparing model results with observed data

#### Get the data and visualize it

```python
# Step 1 - get the data
import numpy as np     # because it's numpy
import pandas as pd    # data library
import matplotlib.pyplot as plt

df = pd.read_csv( "https://people.sc.fsu.edu/~jburkardt/datasets/regression/x06.txt", skiprows=37, 
                 header=None, names=['Index','Age',"Temp","Length"], delim_whitespace=True )

print( df.columns.values )   # print the column headers to the console
print(df.shape)

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.plot( df['Age'].values,df['Length'].values, "+", color="green")
plt.title( "Length vs. Age")

plt.subplot(1,2,2)
plt.plot( df['Temp'].values,df['Length'].values, "o", color='red')
plt.title( "Length vs. Temp")
plt.show()
```

#### Let's also plot in in 3D

```python
%matplotlib notebook    # magic term to enable interactive graphics in the notebook

from mpl_toolkits.mplot3d import Axes3D

# create matplotlib 3d axes
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111,projection='3d')  # Axes3D(fig, azim=-115, elev=15)

# plot data points - points over the HP are white, points below are black
ax.scatter(df['Age'], df['Temp'],df['Length'])
# set axis labels
ax.set_xlabel('Age')
ax.set_ylabel('Temp')
ax.set_zlabel('Length')

plt.show()
```

#### Run the regression

In this case, we will use the statsmodel package.  We will develop several alternative models and compare results.  The model used are:

1. $ Length = a + b Age $
2. $ Length = a + b Age + c Age^2 $
3. $ Length = a + b Age + c Temp $
2. $ Length = a + b Age + c Age^2 + d Temp $


```python
# Step 2 - run regression
import statsmodels.api as sm # import statsmodels 

####### Model 1: y=a+bX ########

XObs = df[ ['Age' ] ] # pull out data columns
yObs = df[ 'Length' ]

XObs = sm.add_constant( XObs )    # include an intercept term
print( XObs.columns.values)

# create and fit the model
model = sm.OLS( yObs, XObs )      # Note the difference in argument order
est = model.fit()
print("\nMODEL: Length = a + b*Age")
print( est.summary() )

######## Model 2: y=a + bX + cX^2 #########

# add an "Age^2" column
Age = df['Age']
Age2 = pd.DataFrame( {"Age2" : Age*Age });
XObs = XObs.join(Age2)

model = sm.OLS( yObs, XObs )      # Note the difference in argument order
est = model.fit()
print("\nMODEL: Length = a + b*Age + c*Age^2")
print( est.summary() )

######## Model 3 y=a + bX + cZ #########

XObs = df[ ['Age','Temp'] ]     # make them numpy arrays for convenience
XObs = sm.add_constant( XObs )    # include an intercept term

model = sm.OLS( yObs, XObs )      # Note the difference in argument order
est = model.fit()
print("\nMODEL: Length = a + b*Age + c*Temp")
print( est.summary() )

######## Model 4 y = a + bX + cX^2 + dZ #########

XObs = df[ ['Age','Temp'] ]     # make them numpy arrays for convenience
XObs = sm.add_constant( XObs )    # include an intercept term
XObs = XObs.join(Age2)

model = sm.OLS( yObs, XObs )      # Note the difference in argument order
est = model.fit()
print("\nMODEL: Length = a + b*Age + c*Age^2 + d*Temp")
print( est.summary() )
```

#### Repeat, but this time with sklearn package

```python
import numpy as np
from sklearn.linear_model import LinearRegression

def PrintResults(model,r2,coefs,intercept):
    print("\n\n", model)
    print("   r2: {:.3}".format(r2))
    print("   coefs:", coefs)
    print("   intercept: {:.4}".format(intercept))

    
####### Model 1: y=a+bX ########

XObs = df[ ['Age' ] ] # make them numpy arrays for convenience
yObs = df[ 'Length' ]

model = LinearRegression().fit(XObs, yObs)
PrintResults("MODEL: Length = a + b*Age", model.score(XObs,yObs),model.coef_, model.intercept_)

######## Model 2: y=a + bX + cX^2 #########

# add an "Age^2" column
Age = df['Age']
Age2 = pd.DataFrame( {"Age2" : Age*Age });
XObs = XObs.join(Age2)

model = LinearRegression().fit(XObs, yObs)
PrintResults("MODEL: Length = a + b*Age + c*Age^2", model.score(XObs,yObs),model.coef_, model.intercept_)

######## Model 3 y=a + bX + cZ #########

XObs = df[ ['Age','Temp'] ]     # make them numpy arrays for convenience
model = LinearRegression().fit(XObs, yObs)
PrintResults("MODEL: Length = a + b*Age + c*Temp", model.score(XObs,yObs),model.coef_, model.intercept_)

######## Model 4 y = a + bX + cX^2 + dZ #########

XObs = df[ ['Age','Temp'] ]     # make them numpy arrays for convenience
XObs = XObs.join(Age2)

model = LinearRegression().fit(XObs, yObs)
PrintResults("MODEL: Length = a + b*Age + c*Age^2 + d*Temp", model.score(XObs,yObs),model.coef_, model.intercept_)
```

## Logistic Regression

For this example, we will use the Iris dataset that is available online - it describes various attributes of three iris varieties.  Our task is to develop a classifiaction model that will give the probabilities of being in one of iris classes.

We will use the sklearn package for this - it provides both the data and a function for conducting a logistic regression.

We will do two versions - one that uses a single attribute to develop a classifer, and a second version that builds a 2-D classifier.

Our basic steps are:
1) get some data
2) create and fit a model
3) report results

We will start with the 1D, single attribute model

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt

iris = load_iris()
print( iris.feature_names )

# get one column of data - the first (Sepal length)
X = iris.data[:,:1]    # 2D array - rows=observation, cols = iris attributes
Y = iris.target        # an array of the correct classification (iris variety) for each observation

# make and fit a model
logisticModel = LogisticRegression()
logisticModel.fit(X,Y)

# generate prediction 
predProbs = logisticModel.predict_proba(X)[:,0]

plt.figure(figsize=(8, 6))
plt.plot(X,Y,'+')
plt.plot(X,predProbs,'o')
plt.xlabel( "Sepal Length (cm)")
plt.ylabel( "Probability of being Setosa variety (Class I)")
plt.show()

# output model coefficients - because there is one input, we will have one model coefficent (+ intercept)
print( logisticModel.intercept_)
print( logisticModel.coef_)
```

```python
"""
=========================================================
Logistic Regression 3-class Classifier
=========================================================

Show below is a logistic-regression classifiers decision boundaries on the
'iris' dataset. The datapoints are colored according to their labels.

"""
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt

iris = load_iris()
print( iris.feature_names )
#print( iris)

X = iris.data[:,:2]    # 2D array - rows=observation, cols = iris attributes
Y = iris.target        # an array of the correct classification for each observation

logisticModel = LogisticRegression( C=1e5)
logisticModel.fit(X,Y)


for i in range(1,5):
    data = X[-i,:].reshape(1,-1)  # get a single observation; reshape makes it a 2D array
    prediction = logisticModel.predict(data)
    print( 'Data: ', data )
    print( '  Predicted class {}, real class {}'.format( prediction,iris.target[-i]))
    print( '  Probabilities for each class from 0 to 2: {}'.format(logisticModel.predict_proba([data[-1]])))


# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, x_max]x[y_min, y_max].
h=0.02
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = logisticModel.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(figsize=(12, 6))
plt.figure(1, figsize=(4, 3))
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

# Plot also the training points
plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolors='k', cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())

plt.show()
```
