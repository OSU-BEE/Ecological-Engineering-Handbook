In this section, we build on the techniques presented in [Empirical Models and Linear Regression](./Linear-Regression.html), and extend those techniques to consider fitting models with **polynomial functional forms**.  Because these polynomials may contain higher-order terms (e.g. $ \beta_2(x^2),\beta_3(x^3) $, they can effectively fit data that does not generally fall along a simple line.

First, let's look at models containing a relationship between a single explanatory variable and a response variable that is described by a more complex relationship than just a line.  What are some examples of such a relationship?

In the figures below, Human Population Size over Time is shown on the left, and Bacterial Growth as a function of Temperature on the right.  In both cases, it is pretty clear that a simple line will not be able to fit the data effectively - we need higher-order terms in our polynomial to capture the curvature associated with each dataset.

In the first example â€“ global human population (the response variable) as a function of time (the predictor variable), we see that the data clearly does not follow a line, and trying to fit a line to this data would lead to poor model performance.  Instead, because the data appears to follow a simple curve, a more appropriate model might include an additional second-order term, e.g.

$ population = \alpha + \beta_1(Year) + \beta_2(Year^2) $

In the second example, we see two bacterial population growth rates that vary as a function of temperature - growth rates in both cases are small at low temperatures, increase as temperature increases to a moderate level, and the go down as temperature get too hot and reduces growth rates. Again this relationship is nonlinear, and trying to fit a simple line to this data would lead to a very poor model. A single second-order term is probably insufficient, as the curvature of the data is both up and down, depending on the temperature.  In this case, a third-order polynomial would likely provide a good fit to the data.  

$ growthRate = \alpha + \beta_1(Temperature) + \beta_2(Temperature^2) + \beta_3(Temperature^3) $

Let's start with developing a World Population model.  In this case, when plotting the data over time (1600-2000), the curvature is only up, which we can capture with a second-order term.  We'll also develop a Simple Line model as well for comparison purposes.  For this exercise, our two models are:

Model 1 (Simple Line): $ population = \alpha + \beta_1(Year) $

Model 2 (Second-order Polynomial): $ population = \alpha + \beta_1(Year) + \beta_2(Year^2) $

Python code for this example is provided below.  We will use numpy's `polyfit()` function for fitting both models.   


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# read the data
df = pd.read_excel("http://explorer.bee.oregonstate.edu/Topic/Modeling/Data/WorldPopulation.xlsx")

# pull out columns of interest
year = df['year']
population = df['World Population']

# define two model functions
model_0 = lambda a, b1, x: a + b1*x
model_1 = lambda a, b1, b2, x: a + b1*x + b2*x*x

# fit models.
# Model 0 is a first-order polynomial (line)
(b1,a)= np.polyfit(year,population,1)  # the '1' indictates we want to fit a first-order polynomial (line)
yModeled_0 = model_0(a,b1,year)

# Model 1 is a second order polynomial
(b2,b1,a)= np.polyfit(year,population,2)  # this time, fit a second-order polynomial
yModeled_1 = model_1(a,b1,b2,year)

# get r2 values
mean = population.mean()

St = np.sum(np.power((population-mean),2))
Sr = np.sum(np.power((population-yModeled_0),2))
r2_0 = (St-Sr)/St     # coefficient of determination

St = np.sum(np.power((population-mean),2))
Sr = np.sum(np.power((population-yModeled_1),2))
r2_1 = (St-Sr)/St     # coefficient of determination

# output results
print( "Model 0: a={:.3}, b1={:.3}, r2={:.3}".format(a,b1, r2_0))
print( "Model 1: a={:.3}, b1={:.3}, b2={:.3}, r2={:.3}".format(a,b1,b2,r2_1))

# plot the results
plt.figure(figsize=(12,6))
plt.scatter(df['year'], df['World Population'], label='Actual Population')
plt.plot(year, yModeled_0, color='red', label='Model 0 (First-order)')
plt.plot(year, yModeled_1, color='green', label='Model 1 (Second-order)')
plt.title('World Population')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.show()
```

### Interpreting these results

Inspecting the resulting plot, we can visaully see that the second-order polynomial does a significantly better (but imperfect) job of fitting the observational data.   This is unsurprising, as the additional second order term in the model provides more flexibility in fitting the data.

Additional, let's compare the r<sup>2</sup> values for the two models.  The first-order model's r<sup>2</sup> value is 0.63, indicating that a simple line relationship can explain 63 percent of the variability observed in the data.  For the second-order polynomial model, the r<sup>2</sup> value increases to 0.88, indicating that the addition of the second-order term accounted for an additional 25 percent of the observed variability (0.88-0.63=0.25)
