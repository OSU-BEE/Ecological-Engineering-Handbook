# Common Mathematical Functions and Their Shapes


Below are some common mathematical functions and their from for different parameterizations


## Power Function
The Power function has two parameters and is of the form

$ y=ax^b $


```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-colorblind')
plt.figure( figsize=(12,6))

X = np.arange(0,10,0.001)

def theModel(a,b,X):
    return a*X**b

a = 1; b = 2
plt.plot(X,theModel(a,b,X),color='red', label='a={}, b={}'.format(a,b))

a = 5; b=2
plt.plot(X,theModel(a,b,X),color='green', label='a={}, b={}'.format(a,b))

a = 1; b=3
plt.plot(X,theModel(a,b,X),color='blue', label='a={}, b={}'.format(a,b))

a = 1; b=.5
plt.plot(X,theModel(a,b,X),color='brown', label='a={}, b={}'.format(a,b))

a = 300; b=-0.1
plt.plot(X,theModel(a,b,X),color='purple', label='a={}, b={}'.format(a,b))

plt.legend()
plt.show()
```

## Exponential Function
The Power function has two parameters and is of the form

$ y = a * b^X $


```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-colorblind')
plt.figure( figsize=(12,6))

X = np.arange(0,5,0.001)

def theModel(a,b,X):
    return a*np.power(b,X)

a = 1; b = 2
plt.plot(X,theModel(a,b,X),color='red', label='a={}, b={}'.format(a,b))

a = 5; b=2
plt.plot(X,theModel(a,b,X),color='green', label='a={}, b={}'.format(a,b))

a = 1; b=3
plt.plot(X,theModel(a,b,X),color='blue', label='a={}, b={}'.format(a,b))

a = 1; b=.5
plt.plot(X,theModel(a,b,X),color='brown', label='a={}, b={}'.format(a,b))

a = 100; b=0.5
plt.plot(X,theModel(a,b,X),color='purple', label='a={}, b={}'.format(a,b))

plt.legend()
plt.show()
```

## Hyperbolic Functions

$ y = a(\frac{X}{b+X})   $

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-colorblind')
plt.figure( figsize=(12,6))

X = np.arange(0,10,0.001)

def theModel(a,b,X):
    return a*(X/(b+X))

a = 1; b = 2
plt.plot(X,theModel(a,b,X),color='red', label='a={}, b={}'.format(a,b))

a = 5; b=2
plt.plot(X,theModel(a,b,X),color='green', label='a={}, b={}'.format(a,b))

a = 1; b=3
plt.plot(X,theModel(a,b,X),color='blue', label='a={}, b={}'.format(a,b))

a = 1; b=.5
plt.plot(X,theModel(a,b,X),color='brown', label='a={}, b={}'.format(a,b))

plt.legend()
plt.show()
```

## Log Functions

$ y=a*ln(x) + b $

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-colorblind')
plt.figure( figsize=(12,6))

X = np.arange(0,10,0.001)

def theModel(a,b,X):
    return a + b*np.log(X) 

a = 1; b = 0.1
plt.plot(X,theModel(a,b,X),color='red', label='a={}, b={}'.format(a,b))

a = 1; b=1
plt.plot(X,theModel(a,b,X),color='green', label='a={}, b={}'.format(a,b))

a = 1; b=2
plt.plot(X,theModel(a,b,X),color='blue', label='a={}, b={}'.format(a,b))

a = 2; b=2
plt.plot(X,theModel(a,b,X),color='brown', label='a={}, b={}'.format(a,b))

plt.legend()
plt.grid()
plt.show()
```

## Logistic Functions

 $\large y = \frac{1}{a+exp(- \beta X)} $

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-colorblind')
plt.figure( figsize=(12,6))

X = np.arange(-10,10,0.001)

def theModel(a,b,c,X):
    return 1 / (a + np.exp(-b * (X-c))) 

a = 1; b = 0.1;c = 0.0
plt.plot(X,theModel(a,b,c,X),color='red', label='a={}, b={}, c={}'.format(a,b,c))

a = 1; b=1
plt.plot(X,theModel(a,b,c,X),color='green', label='a={}, b={}, c={}'.format(a,b,c))

a = 1; b=2
plt.plot(X,theModel(a,b,c,X),color='blue', label='a={}, b={}, c={}'.format(a,b,c))

a = 2; b=2
plt.plot(X,theModel(a,b,c,X),color='brown', label='a={}, b={}, c={}'.format(a,b,c))

a = 1; b=1; c=5
plt.plot(X,theModel(a,b,c,X),color='purple', label='a={}, b={}, c={}'.format(a,b,c))
                
plt.legend()
plt.grid()
plt.show()
```
