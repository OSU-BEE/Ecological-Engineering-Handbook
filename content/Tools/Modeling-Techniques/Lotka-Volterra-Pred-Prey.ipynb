```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

r = 0.5
a = 0.001
b = 0.0001
km = 0.1

def theModel(Y,t):
    # decode curret state varible values
    H = Y[0]
    P = Y[1]
    
    dH_dt = r*H - a*H*P 
    dP_dt = -km*P + b*H*P
    return [dH_dt, dP_dt]




times = np.arange(0,300.05,0.1)

Y = odeint(theModel,[100,3],times)

# plot results
Prey = Y[:,0]
Predator = Y[:,1]

plt.plot(times, Prey, 'r-', label='Prey')
plt.plot(times, Predator, 'b-', label='Predators')
plt.grid()
plt.legend(loc='best')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('Lotka Volterra Predator/Prey Model')

plt.show()
    
```
